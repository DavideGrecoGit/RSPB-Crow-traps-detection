# Train

```python
from ultralytics import YOLO

model_path = "runs/detect/train_50/weights/best.pt"
data_path = "Datasets/Crow_detection_train/config.yaml"

model = YOLO(model_path)  # load model 

# Train the model
valid_results = model.train(
   data=data_path, epochs=50, imgsz=224,hsv_h=0.015,hsv_s=0.7,hsv_v=0.4, fliplr=0.5, flipud=0.5
)
print(valid_results)
```

# Test

```python
from ultralytics import YOLO

# Load a pretrained YOLOv8n model
model = YOLO("runs/detect/train_50/weights/best.pt")
source = "Datasets/Crow_detection/test/crow_trap"

# Run inference on 'bus.jpg' with arguments
results = model.predict(source, save=False, imgsz=224, conf=0.5)

sum = 0
for result in results:
    if len(result.boxes.cpu().numpy()) > 0:
        sum = sum + 1

print(sum/len(results))
```
