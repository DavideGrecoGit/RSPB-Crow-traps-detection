#### Colored images

```python
from ultralytics import YOLO

model_path = "yolov8n-cls.pt"
data_path = "Datasets/Crow_classify"

model = YOLO(model_path)  # load model 

# Train the model
valid_results = model.train(
   data=data_path, epochs=10, imgsz=224,hsv_h=0.015,hsv_s=0.7,hsv_v=0.4, fliplr=0.5, flipud=0.5
)
print(valid_results)
```

```python
model.val(data="Datasets/Crow_evaluate_classify")
```
